#### Step 1. Define purpose of the ML project.

This project aims to predict fraudulent transactions using synthetic financial datasets sourced from [Kaggle](https://www.kaggle.com/datasets/ealaxi/paysim1/data). Correctly identifying whether a transaction is fraudulent is extremely important as fraudsters aim to empty out the victim's bank account through methods that involve transferring and then immediately cashing out. If fraud occurs and suspicious transactions are not stopped, it can result in life-changing scenarios. Correctly identifying fraudulent transactions will improve the customer's perception of the bank. This project will be an end-to-end pipeline and uploaded to the author's [Github](https://github.com/KeithChenYong/Fraud-Detection). 


<u>List of Attributes</u> 
| Attribute       | Description                                                                                          |
|-----------------|------------------------------------------------------------------------------------------------------|
| `step`            | Maps a unit of time in the real world. In this case, 1 step is 1 hour of time. Total steps 744 (30 days simulation). |
| `type`            | CASH-IN, CASH-OUT, DEBIT, PAYMENT, and TRANSFER.                                                     |
| `amount`          | Amount of the transaction in local currency.                                                         |
| `nameOrig`        | Customer who started the transaction.                                                               |
| `oldbalanceOrg`   | Initial balance before the transaction.                                                             |
| `newbalanceOrig`  | New balance after the transaction.                                                                  |
| `nameDest`        | Customer who is the recipient of the transaction.                                                   |
| `oldbalanceDest`  | Initial balance recipient before the transaction. Note that there is no information for customers that start with M (Merchants). |
| `newbalanceDest`  | New balance recipient after the transaction. Note that there is no information for customers that start with M (Merchants). |
| `isFraud`         | Transactions made by the fraudulent agents inside the simulation. Aiming to profit by taking control of customers' accounts and trying to empty the funds by transferring to another account and then cashing out of the system. |
| `isFlaggedFraud`  | Controls massive transfers from one account to another and flags illegal attempts. An illegal attempt in this dataset is an attempt to transfer more than 200,000 in a single transaction. |

<br><br>Information provided by Kaggle.<br>
- Transactions which are detected as fraud are cancelled, so for fraud detection these columns (oldbalanceOrg, newbalanceOrig, oldbalanceDest, newbalanceDest ) must not be used.

```python
# Expanding the list that's flagged as fraud for potentially suspicious transaction before/after
def expansion(check_df, x=0):
    
    check_list = check_df.index.tolist()
    expand_list = []
    
    for i in check_list:
        expand_list.append(i)
        if i + 1 < len(check_df):
            expand_list.append(i + 1)
        if x == 1 and i - 1 >= 0:
            expand_list.append(i - 1)

    full_check_fraud = pd.DataFrame(expand_list, columns=['Index'])
    full_check_fraud = full_check_fraud.drop_duplicates().sort_values(by='Index')
    full_check_fraud = full_check_fraud['Index'].tolist()
    full_check_fraud = original_df.loc[full_check_fraud] 
    
    return full_check_fraud
```

#### Step 6. Partition the data

#### Customised Stratified Sampling Algorithm
- Relies on undersampling the data that is not in a class of interest. 
- More effective when the features are highly correlated with the target variable.

1. **Divide the available data into two sets (strata):**
   - a. all samples of the class of interest (set A).
   - b. all other samples (set B).
2. **Construct the training set:**
   - a. randomly select 50% of the samples in set A.
   - b. add an equal number of samples from set B.
3. **Construct the test set:**
   - a. select the remaining 50% of samples from set A.
   - b. add enough samples from set B to restore the original ratio from the overall data set.


```python
# Decision 3
# Step 1: Divide the data into two sets
set_A = df[df['isFraud'] == 1]
set_B = df[df['isFraud'] == 0]

# Step 2: Construct the training set
train_A = set_A.sample(frac=0.8, random_state=42)
train_B = set_B.sample(n=len(train_A), random_state=42)  # n is the number of samples in train_A
training_set = pd.concat([train_A, train_B])

# Step 3: Construct the test set
test_A = set_A.drop(train_A.index)
ratio_B = len(set_B) / len(df)
required_B_samples = int((len(test_A) / (1 - ratio_B)) * ratio_B) - len(test_A)
test_B = set_B.drop(train_B.index).sample(n=required_B_samples, random_state=42)
test_set = pd.concat([test_A, test_B])

print("Training Set Composition:\n", training_set['isFraud'].value_counts(normalize=True) * 100)
print("\nTest Set Composition:\n", test_set['isFraud'].value_counts(normalize=True) * 100)
```

<a id="choose"></a>
#### Step 7. Choose the ML technique

1. Logistic Regression. Linear relationship and interpretability.
2. Random Forest. Non-linear relationship and Robustness to Overfitting.
3. XGBoost. Handling Imbalanced Data and Boosting Technique

```python
models = {
    'Logistic Regression': LogisticRegression(penalty='l2', C=1.0, max_iter=100),
    'Random Forest': RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=None),
    'XGBoost': XGBClassifier(n_estimators=100, learning_rate=1.0, max_depth=None, eval_metric='logloss')
}

X = training_set.drop('isFraud', axis=1)  # all columns except 'isFraud'
y = training_set['isFraud']

kf = KFold(n_splits=5, shuffle=True, random_state=42)

results = {name: {'Accuracy': [], 'Recall': [], 'Precision': []} for name in models}

for name, model in models.items():
    for train_index, val_index in kf.split(X):
        X_train, X_val = X.iloc[train_index], X.iloc[val_index]
        y_train, y_val = y.iloc[train_index], y.iloc[val_index]
#         print(X_train.shape, X_val.shape, y_train.shape, y_val.shape)
        
        model.fit(X_train, y_train)
        predictions = model.predict(X_val)
        
        results[name]['Accuracy'].append(accuracy_score(y_val, predictions))
        results[name]['Recall'].append(recall_score(y_val, predictions))
        results[name]['Precision'].append(precision_score(y_val, predictions))

# Display results for each model
for name in models:
    print(f"\n{name} Results:")
    print("Average Accuracy:", sum(results[name]['Accuracy']) / len(results[name]['Accuracy']))
    print("Average Recall:", sum(results[name]['Recall']) / len(results[name]['Recall']))
    print("Average Precision:", sum(results[name]['Precision']) / len(results[name]['Precision']))
```

<a id="use"></a>
#### Step 8. Use the ML technique

All models seem to perform well on the training dataset, which consists of 80% fraudulent records. By using a Customized Stratified Sampling Algorithm, we used the remaining 20% of fraudulent records and added non-fraudulent data until the original ratio was restored for the test dataset.

```python
X_test = test_set.drop('isFraud', axis=1)  # all columns except 'isFraud'
y_test = test_set['isFraud']

results_test = {name: {'Accuracy': [], 'Recall': [], 'Precision': []} for name in models}

for name, model in models.items():
    # Re-train the model on the entire training set
    model.fit(X, y)
    predictions = model.predict(X_test)
    
    results_test[name]['Accuracy'].append(accuracy_score(y_test, predictions))
    results_test[name]['Recall'].append(recall_score(y_test, predictions))
    results_test[name]['Precision'].append(precision_score(y_test, predictions))

# Display results for each model
for name in models:
    print(f"\n{name} Results:")
    print("Average Accuracy:", sum(results_test[name]['Accuracy']) / len(results_test[name]['Accuracy']))
    print("Average Recall:", sum(results_test[name]['Recall']) / len(results_test[name]['Recall']))
    print("Average Precision:", sum(results_test[name]['Precision']) / len(results_test[name]['Precision']))
```

<a id="interpret"></a>
#### Step 9. Interpret the results

XGBoost and Logistic Regression perform quite well on the test set. Using the coefficient data from Logistic Regression and feature importances from XGBoost, we can determine which features most significantly help the models.

```python
feature_names = X.columns

for name, model in models.items():
    print(f"\n{name} Model Feature Importance/Coef:")
    
    if name == 'Random Forest' or name == 'XGBoost':
        # Extract feature importances
        importances = model.feature_importances_
        
        # Create a bar chart for the feature importances
        indices = np.argsort(importances)[::-1]
        plt.figure()
        plt.title(f"Feature Importances in {name}")
        plt.bar(range(X.shape[1]), importances[indices], align="center")
        plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
        plt.xlim([-1, X.shape[1]])
        plt.show()
    
    elif name == 'Logistic Regression':
        # Extract coefficients
        coefs = model.coef_[0]
        
        # Create a bar chart for the coefficients
        indices = np.argsort(coefs)[::-1]
        plt.figure()
        plt.title(f"Coefficients in {name}")
        plt.bar(range(X.shape[1]), coefs[indices], align="center")
        plt.xticks(range(X.shape[1]), feature_names[indices], rotation=90)
        plt.xlim([-1, X.shape[1]])
        plt.show()
```

For both XGBoost and Logistic Regression, which are the top two performing algorithms, the two feature-engineered columns provide the most useful information for the algorithms.
